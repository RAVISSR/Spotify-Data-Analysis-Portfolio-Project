```python
corr_df = df_tracks.drop(['key','mode'],axis = 1 ).corr(method = 'pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot = True,fmt='.1g',vmin=-1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x="energy", color='r').set(title="Loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x="acousticness", color='r').set(title="popularity vs acousticness correlation")
```

```python
total_duration = df_tracks.duration_ms
fig_dimensions = (18,7)
fig, ax = plt.subplots(figsize = fig_dimensions)
fig = sns.barplot(x = df_tracks.time_signature, y = total_duration, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

```python
years = df_tracks.time_signature
sns.set_style(style="whitegrid")
fig_dimensions = (10,5)
fig, ax = plt.subplots(figsize= fig_dimensions)
fig = sns.lineplot(x=years, y=total_duration,ax=ax).set(title="Years vs Duration")
plt.xticks(rotation=60)
```

```python
plt.title("Duration of songs in different genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y="genre", x= "duration_ms",data=df_tracks)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = df_tracks.sort_values("popularity", ascending = False).head(10)
sns.barplot(y='genre', x='popularity', data= famous).set(title="Top 5 Genres by popularity")
```
